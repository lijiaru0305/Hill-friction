```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size':22})

import warnings
warnings.filterwarnings('ignore')
```

```python
fig,axs = plt.subplots(3,1, sharex=True, sharey=False, figsize=(12,10))
fig.subplots_adjust(hspace=0.2)

def min180To180(val):
    val = val%(2*np.pi)
    lindex = val<-np.pi
    rindex = val>+np.pi
    val = val + 2*np.pi*lindex - 2*np.pi*rindex
    return (val/np.pi)


def min0To360(val):
    val = val%(2*np.pi)
    return (val/np.pi)
    
p12_data = np.load('./data/res-res/m5K33.npy')
#p12_data = np.load('./data/res-res/orbit_time_000.npy')

t = p12_data[:,0,0]
l0 = p12_data[:,0,5]
l1 = p12_data[:,1,5]
pm0 = p12_data[:,0,3] + p12_data[:,0,4]
pm1 = p12_data[:,1,3] + p12_data[:,1,4]
p = 13
q = 1
ang0 = (p+q)*l1 - p*l0 - q*pm0
ang1 = (p+q)*l1 - p*l0 - q*pm1

phi = l1-pm1


dl = l0-l1
ind_safe = np.abs(min180To180(dl))>(1/180.)
ind_conj = np.arange(len(t)-2)[ (dl[:-2]<0) * (dl[2:]>0) * (np.abs(dl[1:-1])<0.03) ]+1
print(len(ind_conj)*14)
    
axs[2].scatter(t[ind_safe], min0To360(ang1[ind_safe]),s=0.0002,rasterized=True)
axs[1].scatter(t[ind_conj], min0To360(phi[ind_conj]),s=2,rasterized=True)
    
nskip = 10
dphi = phi[ind_conj+nskip] - phi[ind_conj-nskip]
de = p12_data[ind_conj+nskip,1,2] - p12_data[ind_conj-nskip,1,2]

axs[0].scatter(t[ind_conj],de,s=3,rasterized=True)





p12_data = np.load('./data/res-res/m5K34.npy')

t = p12_data[:,0,0]
l0 = p12_data[:,0,5]
l1 = p12_data[:,1,5]
pm0 = p12_data[:,0,3] + p12_data[:,0,4]
pm1 = p12_data[:,1,3] + p12_data[:,1,4]
p = 13
q = 1
ang0 = (p+q)*l1 - p*l0 - q*pm0
ang1 = (p+q)*l1 - p*l0 - q*pm1

phi = l1-pm1


dl = l0-l1
ind_safe = np.abs(min180To180(dl))>(1/180.)
ind_conj = np.arange(len(t)-2)[ (dl[:-2]<0) * (dl[2:]>0) * (np.abs(dl[1:-1])<0.03) ]+1
print(len(ind_conj)*14)
    
axs[2].scatter(t[ind_safe], min0To360(ang1[ind_safe]),s=0.0002,rasterized=True)
axs[1].scatter(t[ind_conj], min0To360(phi[ind_conj]),s=2,rasterized=True)
    
nskip = 10
dphi = phi[ind_conj+nskip] - phi[ind_conj-nskip]
de = p12_data[ind_conj+nskip,1,2] - p12_data[ind_conj-nskip,1,2]

axs[0].scatter(t[ind_conj],de,s=3,rasterized=True)

axs[0].set_xlim([0,1e5])
axs[2].set_ylim([0,2])
axs[1].set_ylim([0,2])
axs[0].set_ylim([-0.03,0.03])

axs[0].set_ylabel(r'$\Delta e$')
axs[1].set_ylabel(r'$\phi/\pi$')
axs[2].set_ylabel(r'$\theta/\pi$')

axs[2].set_xlabel(r'Time [$P_1$]')

axs[0].scatter([],[],color='tab:blue',label=r'$K=3.3$')
axs[0].scatter([],[],color='tab:orange',label=r'$K=3.4$')
axs[0].legend(ncol=2,loc=9,frameon=False)
axs[0].set_title('Two $\mu=10^{-5}$ Runs')

plt.savefig('./pics/pics-island-m5-rev.pdf',bbox_inches='tight',transparent=True)
plt.savefig('./pics/pics-island-m5-rev.png',bbox_inches='tight',transparent=True)
```
