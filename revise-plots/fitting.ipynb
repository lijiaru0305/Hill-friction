```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size':22})

import warnings
warnings.filterwarnings('ignore')
```

```python
from scipy.optimize import curve_fit

N = 100
data = np.load('./data/1p/no-gas-m1e-%1d.npy'%(6))
K = data[:,2]
Tinst = data[:,3]
edges = np.linspace(2.616161616161616,3.47,N+1)
Kl = np.zeros(N)
Tl = np.zeros(N)
for i in range(N):
    ind_i = (K>edges[i]) * (K<edges[i+1]) 
    Kl[i] =  0.5*(edges[i]+edges[i+1])
    Tl[i] =  np.mean(Tinst[ind_i])
X_fit = Kl
Y_fit = np.log10(Tl)
def Tins_fit(K,b,c):
    return b*(K-2.616161616161616)+c
popt, pcov = curve_fit(Tins_fit, X_fit, Y_fit)
print(popt,np.sqrt(np.diag(pcov))*3)
plt.plot(X_fit,Y_fit)

data = np.load('./data/1p/no-gas-m1e-%1d.npy'%(5))
K = data[:,2]
Tinst = data[:,3]
edges = np.linspace(2.595959595959596,3.47,N+1)
Kl = np.zeros(N)
Tl = np.zeros(N)
for i in range(N):
    ind_i = (K>edges[i]) * (K<edges[i+1])
    Kl[i] =  0.5*(edges[i]+edges[i+1])
    Tl[i] =  np.mean(Tinst[ind_i])
X_fit = Kl
Y_fit = np.log10(Tl)
def Tins_fit(K,b,c):
    return b*(K-2.595959595959596)+c
popt, pcov = curve_fit(Tins_fit, X_fit, Y_fit)
print(popt,np.sqrt(np.diag(pcov))*3)
plt.plot(X_fit,Y_fit)

data = np.load('./data/1p/no-gas-m1e-%1d.npy'%(4))
K = data[:,2]
Tinst = data[:,3]
edges = np.linspace(2.5353535353535355,3.43,N+1)
Kl = np.zeros(N)
Tl = np.zeros(N)
for i in range(N):
    ind_i = (K>edges[i]) * (K<edges[i+1])
    Kl[i] =  0.5*(edges[i]+edges[i+1])
    Tl[i] =  np.mean(Tinst[ind_i])
X_fit = Kl
Y_fit = np.log10(Tl)
def Tins_fit(K,b,c):
    return b*(K-2.5353535353535355)+c
popt, pcov = curve_fit(Tins_fit, X_fit, Y_fit)
print(popt,np.sqrt(np.diag(pcov))*3)
plt.plot(X_fit,Y_fit)

data = np.load('./data/1p/no-gas-m1e-%1d.npy'%(3))
K = data[:,2]
Tinst = data[:,3]
edges = np.linspace(2.393939393939394,3.69,N+1)
Kl = np.zeros(N)
Tl = np.zeros(N)
for i in range(N):
    ind_i = (K>edges[i]) * (K<edges[i+1])
    Kl[i] =  0.5*(edges[i]+edges[i+1])
    Tl[i] =  np.mean(Tinst[ind_i])
X_fit = Kl
Y_fit = np.log10(Tl)
def Tins_fit(K,b,c):
    return b*(K-2.393939393939394)+c
popt, pcov = curve_fit(Tins_fit, X_fit, Y_fit)
print(popt,np.sqrt(np.diag(pcov))*3)
plt.plot(X_fit,Y_fit)
```
